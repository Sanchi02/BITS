```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open ('wine-enc-time', 'rb') as fp:
    enc_time = pickle.load(fp)
    
with open ('wine-enc-acc', 'rb') as fp:
    enc_acc = pickle.load(fp)
    
with open ('wine-enc-comp', 'rb') as fp:
    enc_comp = pickle.load(fp)
    
with open ('wine-time', 'rb') as fp:
    time = pickle.load(fp)
    
with open ('wine-acc', 'rb') as fp:
    acc = pickle.load(fp)
    
with open ('wine-comp', 'rb') as fp:
    comp = pickle.load(fp)
```

```python
print("Avg batch time for encrypted data {:0.5f}s".format(sum(enc_time)/len(enc_time)))
print("Avg batch time for un-encrypted data {:0.5f}s".format(sum(time)/len(time)))
```

```python
m = min(len(enc_time),len(time))
cnt = np.arange(1, m+1, 1)


plt.plot( cnt,enc_time[:m],label="Encrypted")
plt.plot( cnt,time[:m],label="Un-encrypted")


axes = plt.gca()
#axes.set_ylim([-1.4,1.4])

plt.legend(loc="center right")
plt.xlabel('batch')
plt.ylabel('time (s)')
plt.title('Time taken per batch for training phase')
plt.grid(True)
plt.show()
```

```python
m = min(len(enc_acc),len(time))
cnt = np.arange(1, m+1, 1)

print(len(enc_acc))
print(len(acc))

plt.plot( cnt,enc_acc[:m],label="Encrypted")
plt.plot( cnt,acc[5:m+5],label="Un-encrypted")

axes = plt.gca()
#axes.set_ylim([-1.4,1.4])

plt.legend(loc="lower right")
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('Accuracy through training phase')
plt.grid(True)
plt.show()
```

```python
nn_model_neurons = [16,24,32,48,64,84]

m = min(len(enc_comp),len(time))

plt.plot( nn_model_neurons,enc_comp[:m],label="Encrypted")
plt.plot( nn_model_neurons,comp[:m],label="Un-encrypted")

axes = plt.gca()
#axes.set_ylim([-1.4,1.4])

plt.legend(loc="upper right")
plt.xlabel('neurons')
plt.ylabel('accuracy')
plt.title('Accuracy with varying model complexity')
plt.grid(True)
plt.show()
```
