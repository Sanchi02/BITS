```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open ('fashion-enc-time', 'rb') as fp:
    enc_time = pickle.load(fp)
    
with open ('fashion-enc-acc', 'rb') as fp:
    enc_acc = pickle.load(fp)

    
with open ('fashion-time', 'rb') as fp:
    time = pickle.load(fp)
    
with open ('fashion-acc', 'rb') as fp:
    acc = pickle.load(fp)
    
```

```python
print("Avg batch time for encrypted data {:0.5f}s".format(sum(enc_time)/len(enc_time)))
print("Avg batch time for un-encrypted data {:0.5f}s".format(sum(time)/len(time)))
```

```python
m = min(len(enc_time),len(time))
cnt = np.arange(1, m+1, 1)


plt.plot( cnt,enc_time[:m],label="Encrypted")
plt.plot( cnt,time[:m],label="Un-encrypted")


axes = plt.gca()
#axes.set_ylim([-1.4,1.4])

plt.legend(loc="center right")
plt.xlabel('batch')
plt.ylabel('time (s)')
plt.title('Time taken per batch for training phase')
plt.grid(True)
plt.show()
```
