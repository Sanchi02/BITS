```python
levels = 3
nodesNames = ['root']
n = 1
fileName = 'L9-Full.txt'
for i in range(1,levels):
    n = n*2
    tmp = ""
    for j in range(1,n+1):
        tmp = 'l'+str(i)+'-'+str(j)
        nodesNames.append(tmp)
nodesNames
```

```python
links = []
numNodes = (2**levels) - 1
for i in range(numNodes//2):
    links.append([nodesNames[i],nodesNames[(i*2)+1],'10Mbps','1','1ms','100'])
    links.append([nodesNames[i],nodesNames[(i*2)+2],'10Mbps','1','1ms','100'])
print(links)
```

```python
routers = []
for n in nodesNames:
    routers.append([n,'NA','20','20'])
print(routers)
```

```python
import pandas as pd

filePtr = open(fileName,'a')
filePtr.write("router\n")
filePtr.write('\n')
filePtr.close()
routers = pd.DataFrame(routers, columns = ['node', 'city','y','x'])
links = pd.DataFrame(links, columns = ['from', 'to','capacity','metric','delay','queue']) 

routers.to_csv(fileName, header=None, index=None, sep='\t', mode='a')
filePtr = open(fileName,'a')
filePtr.write("\nlink\n")
filePtr.write('\n')
filePtr.close()
links.to_csv(fileName, header=None, index=None, sep='\t', mode='a')
```
